```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile

# Step 2: Load the data from ZIP
with zipfile.ZipFile("API_SP.POP.TOTL_DS2_en_csv_v2_76034.zip") as z:
    print(z.namelist())  # check file name inside
    with z.open("API_SP.POP.TOTL_DS2_en_csv_v2_76034.csv") as f:
        df = pd.read_csv(f, skiprows=4)

# Step 3: Drop unnecessary columns (DO NOT drop 'Country Name')
df.drop(['Indicator Name', 'Indicator Code'], axis=1, inplace=True)

# Step 4: Drop rows with missing values in 2022
df = df.dropna(subset=['2022'])

# Step 5: Sort data based on population in 2022
df_top10 = df.sort_values(by='2022', ascending=False).head(10)

# Step 6: Plot and save the result
plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', data=df_top10)
plt.title("Top 10 Countries by Population in 2022")
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("top10_population_2022.png")
plt.show()

# Step 7: (Optional) Print country names
print("Top 10 Countries of total population in 2022:")
print(df_top10[['Country Name', '2022']])

```

```python
# Step 4: Drop rows with missing values in 2016
df = df.dropna(subset=['2016'])

# Step 5: Sort data based on population in 2016
df_top10 = df.sort_values(by='2016', ascending=False).head(10)

# Step 6: Plot and save the result
plt.figure(figsize=(10, 6))
sns.barplot(x='2016', y='Country Name', data=df_top10)
plt.title("Top 10 Countries by Population in 2016")
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("top10_population_2016.png")
plt.show()

# Step 7: (Optional) Print country names
print("Top 10 Countries of total population in 2016:")
print(df_top10[['Country Name', '2016']])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for male and female population
male_population_data = pd.DataFrame({
    'Country Code': ['USA', 'CHN', 'IND', 'BRA', 'RUS'],
    '2022': [165000000, 720000000, 720000000, 108000000, 74000000]
})

female_population_data = pd.DataFrame({
    'Country Code': ['USA', 'CHN', 'IND', 'BRA', 'RUS'],
    '2022': [168000000, 692000000, 680000000, 106000000, 70000000]
})

# Merge the data
merged_data = pd.merge(male_population_data, female_population_data, on='Country Code', suffixes=('_male', '_female'))

# Create a bar plot (using total population for simplicity, summing male and female)
merged_data['Total_2022'] = merged_data['2022_male'] + merged_data['2022_female']
plt.figure(figsize=(12, 6))
sns.barplot(data=merged_data, x='Country Code', y='Total_2022', palette='viridis')
plt.title('Total Population by Country in 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for male and female population
male_population_data = pd.DataFrame({
    'Country Code': ['USA', 'CHN', 'IND', 'BRA', 'RUS'],
    'Population': [165000000, 720000000, 720000000, 108000000, 74000000]
})

female_population_data = pd.DataFrame({
    'Country Code': ['USA', 'CHN', 'IND', 'BRA', 'RUS'],
    'Population': [168000000, 692000000, 680000000, 106000000, 70000000]
})

# Merge the data
merged_data = pd.merge(male_population_data, female_population_data, on='Country Code', suffixes=('_male', '_female'))

# Reshape data for bar plot (long format)
gender_data = pd.melt(merged_data, id_vars=['Country Code'], 
                      value_vars=['Population_male', 'Population_female'],
                      var_name='Gender', value_name='Population')

# Create bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=gender_data, x='Country Code', y='Population', hue='Gender', palette='viridis')
plt.title('Gender Distribution by Country in 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
